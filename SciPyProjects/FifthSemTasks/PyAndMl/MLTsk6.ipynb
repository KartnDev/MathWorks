```python
from sklearn import linear_model
from sklearn import datasets

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# First example

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = - 5 * x - 100 + rng.randn(50)
plt.scatter(x, y)
plt.show()
```

```python
model = linear_model.LinearRegression(fit_intercept=True)
model.fit(x[:, np.newaxis], y)
xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])
plt.scatter(x, y)
plt.plot(xfit, yfit)
```

```python
print('Accuracy: :', model.score(x[:, np.newaxis], y))
```

We created a linear samples that are noised not so much that can be decrease the accuracy, so we have very good model to predict linear feature
Мы создали линейную выборку которая слабо зашумлена чтобы повлиять на снижение точности, потому мы получили можель с высокой точностью (выше 99)

# Real Example

```python
data = datasets.load_boston()

df = pd.DataFrame(data.data, columns=data.feature_names)
df
```

```python
target = pd.DataFrame(data.target, columns=["MEDV"])
target
```

```python
X = df
y = target["MEDV"]
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X, y)

predictions = lm.predict(X)
predictions[::10] # each 10th
```

```python
print('Accuracy: :', lm.score(X, y))
```

# Better estimator

we take real dataset and try to estimate? bun in real life data cannot be estimated as linear

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
degree = 2
polyreg = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model = polyreg.fit(X, y)

predictions = polyreg.predict(X)
predictions[::10] # each 10th
```

```python
print('Accuracy: :', polyreg.score(X, y))
```

```python
degree = 3
polyreg = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model = polyreg.fit(X, y)

predictions = polyreg.predict(X)
predictions[::10] # each 10th
print('Accuracy: :', polyreg.score(X, y))
```

as we see polinomial regression is better model to solve real-life problem
