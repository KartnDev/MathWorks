```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras as keras
from tensorflow.keras import regularizers
```

```python
def do_vector():
  first_point = np.random.rand(2,) 
  second_point = np.random.rand(2,)
  right_direction = 1 if first_point[0] < second_point[0] else 0
  return np.array([first_point, second_point]).reshape(4, ), right_direction
```

```python
sample = [do_vector() for _ in range(10000)]
sample_vectors = np.array([item[0] for item in sample])
sample_y = np.array([item[1] for item in sample])
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(10, input_shape=(4,)))
model.add(keras.layers.Dense(10, activation='sigmoid', kernel_initializer='he_uniform'))
model.add(keras.layers.Dense(1))
model.compile(loss='mse', optimizer='adam')
```

```python
model.fit(sample_vectors, sample_y, epochs=10)
```

```python
def format_output(output):
  return "Right" if output > 0.5 else "Left"
```

```python
size = 10000
ok_predicts = 0
print_each = 1000
test = [do_vector() for _ in range(size)]
test_vectors = np.array([item[0] for item in test])
test_y = np.array([item[1] for item in test])
predicted = model.predict(test_vectors)
for i, (pred, real) in enumerate(zip(predicted, test_y)):
  if format_output(real) == format_output(pred):
    ok_predicts += 1
  if i % print_each == 0:
    print(f"Real one is {format_output(real)}, predicted: {format_output(pred)}")
```

```python
print(f"Accuracy for test data with size={size} is {ok_predicts/size*100}%")
```
