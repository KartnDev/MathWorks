```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras as keras
from tensorflow.keras import regularizers
```

# Вариант 6
## Определить a, b, c для зависимости y = alog(x + b) + sin(cx)
Диапазон неизвестных параметров [0.1, 1]
Данные генерируются слуайно, х от 0 до 1
Составить у исходя из коеффициентов и х

```python
x = np.array([1.53, 1.3, 1.23, 1.11, 1.04, 0.87, 0.67, 0.59, 0.5])
a = np.random.rand(100)
b = np.random.rand(100)
c = np.random.rand(100)

func = lambda x, i: a[i] * np.log(x + b[i]) + np.sin(c[i] * x)
y = np.array([func(x, i) for i in range(len(a))])
print(y)
```

```python
model = keras.Sequential([
                          keras.layers.Input(shape=(9,)),
                          keras.layers.Dense(18, activation='tanh', kernel_regularizer=regularizers.l2(l2=0.001)),
                          keras.layers.Dense(3, activation='tanh')

])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=['accuracy'])
```

```python
print(np.column_stack((a, b, c)).shape)
print(y.shape)
```

```python
model.fit(y, np.column_stack((a, b, c)), epochs=500);
```

```python
a = 0.17
b = 0.49
c = 0.19
y = np.array([a * np.log(x + b) + np.sin(c * x)])
coefs_pred = model.predict(y)
print(coefs_pred)
```

```python
diff = np.round([[a, b, c]] - coefs_pred, 3)[0]
print(f"Difference between predicted and real is: \na = {diff[0]} \nb = {diff[1]}\nc = {diff[2]}")
```
